```python
# setup preprocessor

preprocessor.set_params(
    vectorizer__n_components=256, # dimensionality of the word vectors
    vectorizer__w2v={
        'iter':128, # number of training iterations of the w2v model
        'min_count':2, # all words with total frequency lower than this will be ignored
    },
)
preprocessor.fit(train_data)
train_data = preprocessor.transform(train_data)
test_data = preprocessor.transform(test_data)
```

```python
# optional
# TA - seg
validation_fraction = 0.1 # validation proportion
mask = np.random.rand(len(train_data)) < (1-validation_fraction)

val_split = train_data.iloc[~mask]
train_split = train_data.iloc[mask]
```

```python
def split_test(vec):
    """
    Change vec1 and vec2 test segment in both data frame to two distinct data frame with columns corresponding to each vec* element.
    :param vec: type of vector; It could be 'vec1' or 'vec2'
    """
    test_split = test_data
    test_ls = []
    for idx, row in test_split.iterrows():
        test_ls.append(row[vec])
        
    test_X_df = pd.DataFrame(test_ls)
    test_y_df = pd.DataFrame(test_split["popularity"])
    test_y = np.array(test_y_df.T.values.tolist()[0])
    return (test_X_df, test_y)

test_X_df_vec1, test_y_vec = split_test('vec_1')
test_X_df_vec2, test_y_vec = split_test('vec_2')
```

```python
def split_Xy(vec):
    """
    Change vec1 and vec2 train and val segment in both data frame to two distinct data frame with columns corresponding to each vec* element.
    :param vec: type of vector; It could be 'vec1' or 'vec2'
    """
    train_ls = []
    for idx, row in train_split.iterrows():
        train_ls.append(row[vec])
        
    train_X_df = pd.DataFrame(train_ls)
    train_y_df = pd.DataFrame(train_split["popularity"])
    train_y = np.array(train_y_df.T.values.tolist()[0])

    val_ls = []
    for idx, row in val_split.iterrows():
        val_ls.append(row[vec])
    val_X_df = pd.DataFrame(val_ls)
    val_y_df = pd.DataFrame(val_split["popularity"])
    val_y = np.array(val_y_df.T.values.tolist()[0])
    
    return (train_X_df, val_X_df, train_y, val_y)

train_X_df_vec1, val_X_df_vec1, train_y_vec, val_y_vec = split_Xy('vec_1')
train_X_df_vec2, val_X_df_vec2, train_y_vec, val_y_vec = split_Xy('vec_2')

train_X_df_vec1.head(1)
```

```python
def convert_dic_for_print(score):
    ret = dict()
    for key in score:
        for key2 in score[key]:
            if key2 == "accuracy":
                ret['ACC'] = score[key][key2]
                continue
            ret[key2+'_'+key] = score[key][key2]
    return ret
```

```python
def convert_dataframe_to_marker(ls):
    df_maker = np.zeros((len(ls)))
    for i in range(len(ls)):
        df_maker[i] = 1
        if ls[i] == 1:
            df_maker[i] = 0
    return df_maker

def plot_model(y, y_hat, vec, model_type, vectype):
    x = vec

    pca = PCA(n_components=2)
    pca.fit(x)
    principalComponents = pca.transform(x)

    principalDf = pd.DataFrame(data = principalComponents
                 , columns = ['PC1', 'PC2'])

    finalDf = pd.concat([principalDf, test_data['popularity']], axis = 1)
    X = finalDf

    marker_ls = convert_dataframe_to_marker(test_data['popularity'])

    color_ls = y_hat

    X1 = X.loc[marker_ls == 1, :]
    color_ls_1 = color_ls[marker_ls == 1]
    plt.figure(figsize=(10,10))

    plt.scatter(X1.iloc[:, 0], X1.iloc[:, 1]
                    , c = color_ls_1
                    , s = 20 ,
                    cmap = 'winter', marker = '<', alpha =1, )

    
    X2 = X.loc[marker_ls == 0, :] 
    color_ls_2 = color_ls[marker_ls == 0]

    plt.scatter(X2.iloc[:, 0], X2.iloc[:, 1]
                    , c = color_ls_2 
                    , s = 20 ,
                    cmap = 'winter', marker = '>', alpha=1)

    pcps = mpatches.Patch(color='lightgreen', label='Predicted Positive Class')
    pcneg = mpatches.Patch(color='blue', label='Predicted Negative Class')

    plt.legend(handles=[pcps, pcneg])
    plt.suptitle('> : Actual positive Class \n < : Actual negative Class')

    plt.title(model_type + ' - ' + vectype)
    plt.xlabel('pca1')
    plt.ylabel('pca2')
    plt.show()
```

```python
def plot_clusters(y, y_hat, vec, model, vectype, model_type):
    
    pca = PCA(n_components=2)
    pca.fit(vec)
    principalComponents = pca.transform(vec)
    principalDf = pd.DataFrame(data = principalComponents
             , columns = ['PC1', 'PC2'])
    pred = model.predict(vec)
    plt.figure(figsize=(12, 12))    
    plt.scatter(principalComponents[:,0], principalComponents[:,1]
            , c=pred, s=20, cmap='viridis');

    plt.title(model_type + ' - ' + vectype)
    plt.xlabel('pca1')
    plt.ylabel('pca2')
    plt.show()
```

```python
mi = 20
cc = 8
KM_model = clustering.KMeans(cc=cc, mi=mi)
KM_model = KM_model.fit(train_X_df_vec1)
y_hat = KM_model.predict(train_X_df_vec1)
plot_clusters(train_y, y_hat, train_X_df_vec1, KM_model, 'vec1', 'KM')
```
